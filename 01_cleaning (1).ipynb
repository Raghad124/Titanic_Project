```python
#titanic data
#phase 1
from google.colab import files
uploaded = files.upload()

#import pandas library for data handling
import pandas as pd
#load titanic dataset
df = pd.read_csv("train.csv")
#show 5 first rows
print(df.head())

#check number of rows and columns
print("shape:", df.shape)
#check data types and column informations
df.info()
#check Missing values in each column
print("Missing values:")
print(df.isnull().sum())
#check duplicate rows
print("duplicate rows:")
print(df.duplicated().sum())

#Function to clean the dataset
def clean_data(df):
    df = df.drop_duplicates()
    #convert survived to category
    df["Survived"] = df["Survived"].astype("category")

    #convert Passenger Class to category
    df["Pclass"] = df["Pclass"].astype("category")

    #Fill with Missing Age values with median
    df["Age"] = df["Age"].fillna(df["Age"].median())
    #Fill  Missing Embarked values with mode
    df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

    #Drop cabin because it has many Missing values
    df = df.drop(columns=["Cabin"])
    #Find 99th percentile of Fare
    cap_value = df["Fare"].quantile(0.99)
    #cap extreme Fare values
    df["Fare"] = df["Fare"].clip(upper=cap_value)
    return df


df = clean_data(df)

# Check that there are no missing values left in any column
assert df.isnull().sum().sum() == 0

#Check that all Age values are non-negative
assert (df["Age"] >= 0).all()

#Check that the number of columns is correct (after dropping 'Cabin')
assert df.shape[1] == 11

#show cleaned dataset
print(df.head())
```
