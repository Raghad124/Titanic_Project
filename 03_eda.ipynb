```python
#Phase3
#Titanic data
from google.colab import files
uploaded = files.upload()
#import for libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("train.csv")
df.head()

numerical_features = ['Age', 'Fare', 'SibSp']

for feature in numerical_features:
    plt.figure(figsize=(8,4))
    sns.histplot(df[feature].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
    plt.show()

    plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title('Fare distribution by Pclass and Survived')
plt.show()

# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation
corr_matrix = numeric_df.corr()

# Get top 10 features correlated with Survived
top_features = corr_matrix['Survived'].abs().sort_values(ascending=False).head(11).index

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df[top_features].corr(), annot=True, cmap='coolwarm')
plt.title('Top 10 features correlated with Survived')
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=0.7)
plt.title('Age vs Fare colored by Survived')
plt.show()

group_summary = df.groupby('Pclass')['Survived'].mean()
print(group_summary)
#it is talking about Titanic data
#Histogram / KDE
#Age: Most passengers were middle-aged; the distribution is right-skewed.
#Fare: A few passengers paid very high fares.
#SibSp: Most passengers had no siblings or spouses on board.
 #Grouped Boxplots
#First-class passengers paid higher fares, and most survivors were from this class, showing the impact of social class on survival.
#Correlation Heatmap
#The most correlated features with survival are Sex, Pclass, and Fare.
 #Scatter plot
#Children had higher survival chances, and passengers with higher fares were more likely to survive.
 #Groupby summary
#Highest survival rate: First class (~62%), lowest: Third class (~24%)

```
