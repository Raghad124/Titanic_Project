```python
from google.colab import files
uploaded = files.upload()

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("train.csv")
print(df.head())
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1) Age distribution
axes[0, 0].hist(df['Age'].dropna(), bins=30)
axes[0, 0].set_title("Age Distribution")

# 2) Fare distribution
axes[0, 1].hist(df['Fare'].dropna(), bins=30)
axes[0, 1].set_title("Fare Distribution")

# 3) Survival by class
sns.barplot(x='Pclass', y='Survived', data=df, ax=axes[1, 0])
axes[1, 0].set_title("Survival Rate by Class")

# 4) Age vs Fare
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, ax=axes[1, 1])
axes[1, 1].set_title("Age vs Fare")

plt.tight_layout()
plt.show()
```
